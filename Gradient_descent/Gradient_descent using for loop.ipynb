```python
import numpy as np
import pandas as pd
```

```python
#Gradient descent for below data
#https://raw.githubusercontent.com/kishore1708/Assignments_Inceptez/main/Gradient_descent/data.csv#

data=pd.read_csv("https://raw.githubusercontent.com/kishore1708/Assignments_Inceptez/main/Gradient_descent/data.csv",header=None)
data.head()
```

```python
data.set_axis(['x','y'],axis=1,inplace=True)
data.head()
```

```python
X=data.loc[:,'x']
Y=data.loc[:,'y']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
#At initial both m and C are 0

m,c=0,0
Y_p=m*X+c

plt.scatter(X,Y)
#plt.plot([min(X),max(X)],[min(y_p),max(y_p)],color='green')
plt.plot(X,Y_p,color='green')
plt.show()
```

```python
#Build the Gradient Descent Model: 
#Step 1: initializing the values for m and c 
m = 0
c = 0

#Step 2: Initialize the learning RAte and iterations 
L = 0.00001
iter_ = 1000

n = len(X)

#For loop - Performing Gradient Descent 
for i in range(iter_):
    Y_p = (m*X)+c #The current prediction value 
    D_m = (-2/n)*sum(X * (Y - Y_p))
    D_c = (-2/n)*sum(Y - Y_p)
    m = m - (L*D_m) 
    c = c - (L*D_c)
    print("The Slope and intercept for iteration:",i+1)
    print("Slope: ",m)
    print("Intercept: ",c)
    print("*"*10)
    
print(f"The final slope is {m:.2f} and intercept is {c:.2f}")
```
