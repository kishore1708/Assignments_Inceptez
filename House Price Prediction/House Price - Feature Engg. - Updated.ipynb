## Feature Engineering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
pd.set_option('display.max_rows', 90)
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
df=pd.concat([train,test],ignore_index=True,axis=0)
df
```

##### Address Missing Values

```python
# For Categorical variables
```

```python
Cat_nafeatures=[feature for feature in df.columns if df[feature].dtype=='O' and df[feature].isna().sum()>=1]
```

```python
for feature in Cat_nafeatures:
    df[feature]=df[feature].fillna('missing')
```

```python
df[Cat_nafeatures].isna().sum()
```

```python
# For Numerical variables
```

```python
Num_nafeatures=[feature for feature in df.columns if df[feature].dtype!='O' and df[feature].isna().sum()>=1 
                and feature!='SalePrice']
```

```python
for feature in Num_nafeatures:
    df[feature+'na']=np.where(df[feature].isnull(),1,0)
```

```python
for feature in Num_nafeatures:
    df[feature]=df[feature].fillna(df[feature].median())
```

```python
df[Num_nafeatures].isna().sum()
```

```python
df.head()
```

##### Temporal variable (Year)

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    df[feature]=df['YrSold']-df[feature]
```

```python
df.head()
```

##### Numeric Variable

```python
# Taking log normal distribuation

num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea','SalePrice']
for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head(10)
```

##### Handling Rare Categorical Feature

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
# We will remove categorical variables that are present less than 1% of the observations

for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df
```

```python
##### Handling categorical columns labels to numerical
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby(feature)['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df
```

```python
feature=[features for features in df.columns if 'na' in features and features!='Functional']
df=df.drop(feature,axis=1)
```

```python
df
```

```python
train_a=df.loc[df['SalePrice'].isna()==False].drop('Id',axis=1)
test_a=df.loc[df['SalePrice'].isna()==True].drop(['SalePrice','Id'],axis=1)
```

```python
test_a.reset_index(drop=True,inplace=True)
```
