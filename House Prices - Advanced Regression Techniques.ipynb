```python
#df.drop(columns=['Id','SalePrice'],axis=1,inplace=True)
df[num_na_col]=df[num_na_col].fillna(df[num_na_col].median())
df[num_na_col].isna().sum()
```

```python
for col in numeric_col:
    low=df[col].quantile(0.25)
    up=df[col].quantile(0.75)
    iqr=up-low
    

    low_outlier = low-1.5*iqr
    up_outlier=up+1.5*iqr
    

    df.loc[(df[col]<low_outlier) | (df[col]>up_outlier),col]=df[col].median()
```
